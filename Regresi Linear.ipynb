```python
import matplotlib.pyplot as pl
import numpy as np

x = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]

a = np.poly1d(np.polyfit(x, y, 1))
b = np.poly1d(np.polyfit(x, y, 2))
c = np.poly1d(np.polyfit(x, y, 3))
uwu = np.linspace(min(x), max(x))

pl.scatter(x, y, color='g')
pl.grid(True)
pl.plot(uwu, a(uwu), 'k')
pl.plot(uwu, b(uwu), 'r', linestyle='dashed')
pl.plot(uwu, c(uwu), 'b', linestyle='dashed')
pl.show()
print("Nilai y linier saat x = 17 adlh %.2f" % a(17))
print("Nilai y derajat 2 saat x = 17 adlh %.2f" % b(17))
print("Nilai y derajat 3 saat x = 17 adlh %.2f" % c(17))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

h = [[x[i]] for i in range(len(x))]
k = [[y[i]] for i in range(len(y))]

def Pol(n, c):
    Poly = PolynomialFeatures(degree=n)
    xpol = Poly.fit_transform(h)
    model = LinearRegression()
    model.fit(xpol, k)
    ypol = model.predict(xpol)
    pl.plot(h, ypol, c)

pl.scatter(x, y, color='g')
Pol(1, 'k')
Pol(2, 'r')
Pol(3, 'b')
pl.grid(True)
```
